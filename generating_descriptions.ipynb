```python
aers_preds = [item for sublist in aers["entity_prediction"] for item in sublist]
```

```python
f_aers_preds = list(filter(lambda x: x["probability"][0] > 0.8, aers_preds))
```

```python
len(f_aers_preds), len(aers_preds)
```

```python
import re

processed_ents = []
for ent in aers_preds:
    if sum(ent["probability"]) / len(ent["probability"]) < 0.8 or min(ent["probability"]) < 0.8:
      continue
    ent["entity_f"] = list(filter(lambda x: len(x) > 2 and x != 'the', ent["entity"]))
    name = re.sub(r'[^\w\s]', '', ' '.join(ent["entity_f"]))
    
    if len(name) > 0:
      processed_ents.append(name)
processed_ents = list(set(processed_ents))
```

```python
processed_ents
```

```python
manual_filtering = ["Captain Crocker", "Lady Brackenstall", "Sir Eustace"]
for ent in processed_ents:
  if "Crocker" in ent or "Brackenstall" in ent or "Eustace" in ent:
    continue
  manual_filtering.append(ent)
```

```python
import networkx as nx
```

```python
G = nx.Graph()
for ent in manual_filtering:
  G.add_node(ent, type='object')
```
